```python
import pandas as pd
from IPython.display import display

file_path = r"C:\Users\vaibh\OneDrive\Desktop\Projects\ETL_Healthcare_Project\healthcare_dataset.xlsx"

df = pd.read_excel(file_path)
```

```python
# ==========================================
# DATA PROFILING
# ==========================================

from IPython.display import display

display(pd.DataFrame({
    "Rows": [df.shape[0]],
    "Columns": [df.shape[1]]
}))
```

```python
display(df.info())
```

```python
duplicates = df[df.duplicated()]

display(duplicates)

display(
    pd.DataFrame({
        "Total Duplicate Rows": [df.duplicated().sum()]
    })
)
```

```python
rows_before = len(df)

display(
    pd.DataFrame({
        "Rows Before": [rows_before]
    })
)
```

```python
df = df.drop_duplicates()
```

```python
rows_after = len(df)

display(
    pd.DataFrame({
        "Rows Before": [rows_before],
        "Rows After": [rows_after],
        "Duplicates Removed": [rows_before - rows_after],
        "Duplicates Remaining": [df.duplicated().sum()]
    })
)
```

```python
# ==========================================
# VALIDATE DUPLICATE REMOVAL
# ==========================================

validation = pd.DataFrame({
    "Validation Check": [
        "Duplicate Records Remaining",
        "Rows After Cleaning"
    ],
    "Result": [
        df.duplicated().sum(),
        len(df)
    ]
})

display(validation)
```

```python
# ==========================================
# FEATURE ENGINEERING
# Create new columns for analysis
# ==========================================

# ------------------------------------------------
# 1. Calculate Length of Stay (in days)
# ------------------------------------------------
df["Length_of_Stay"] = (
    df["Discharge Date"] -
    df["Date of Admission"]
).dt.days

# ------------------------------------------------
# 2. Create Age Groups
# ------------------------------------------------
df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0, 18, 35, 50, 65, 100],
    labels=[
        "Child",
        "Young Adult",
        "Adult",
        "Senior",
        "Elderly"
    ]
)

# ------------------------------------------------
# 3. Create Billing Category
# ------------------------------------------------
df["Billing_Category"] = pd.qcut(
    df["Billing Amount"],
    q=4,
    labels=[
        "Low",
        "Medium",
        "High",
        "Very High"
    ]
)

# ------------------------------------------------
# 4. Create Admission Date Features
# ------------------------------------------------
df["Admission_Year"] = df["Date of Admission"].dt.year
df["Admission_Month"] = df["Date of Admission"].dt.month_name()
df["Admission_Day"] = df["Date of Admission"].dt.day

# ------------------------------------------------
# 5. Create Discharge Date Features
# ------------------------------------------------
df["Discharge_Year"] = df["Discharge Date"].dt.year
df["Discharge_Month"] = df["Discharge Date"].dt.month_name()

# ------------------------------------------------
# 6. Create Stay Category
# ------------------------------------------------
df["Stay_Category"] = pd.cut(
    df["Length_of_Stay"],
    bins=[0, 5, 10, 20, 100],
    labels=[
        "Short Stay",
        "Medium Stay",
        "Long Stay",
        "Extended Stay"
    ]
)

# ==========================================
# Display newly created columns
# ==========================================

display(
    df[[
        "Age",
        "Age_Group",
        "Billing Amount",
        "Billing_Category",
        "Length_of_Stay",
        "Stay_Category",
        "Admission_Year",
        "Admission_Month",
        "Admission_Day",
        "Discharge_Year",
        "Discharge_Month"
    ]].head(10)
)

# ==========================================
# Display total rows and columns
# ==========================================

display(
    pd.DataFrame({
        "Total Rows": [len(df)],
        "Total Columns": [len(df.columns)]
    })
)
```

```python
# ==========================================
# DATA VALIDATION
# Validate the transformed dataset
# ==========================================

from IPython.display import display

# ------------------------------------------
# Missing Values
# ------------------------------------------
missing_values = df.isnull().sum().sum()

# ------------------------------------------
# Duplicate Records
# ------------------------------------------
duplicate_records = df.duplicated().sum()

# ------------------------------------------
# Negative Age
# ------------------------------------------
negative_age = (df["Age"] < 0).sum()

# ------------------------------------------
# Negative Billing Amount
# ------------------------------------------
negative_billing = (df["Billing Amount"] < 0).sum()

# ------------------------------------------
# Negative Length of Stay
# ------------------------------------------
negative_stay = (df["Length_of_Stay"] < 0).sum()

# ------------------------------------------
# Invalid Date Records
# (Discharge Date before Admission Date)
# ------------------------------------------
invalid_dates = (
    df["Discharge Date"] <
    df["Date of Admission"]
).sum()

# ------------------------------------------
# Display Validation Summary
# ------------------------------------------
validation_report = pd.DataFrame({
    "Validation Check": [
        "Missing Values",
        "Duplicate Records",
        "Negative Age",
        "Negative Billing Amount",
        "Negative Length of Stay",
        "Invalid Dates"
    ],
    "Result": [
        missing_values,
        duplicate_records,
        negative_age,
        negative_billing,
        negative_stay,
        invalid_dates
    ],
    "Status": [
        "Pass" if missing_values == 0 else "Fail",
        "Pass" if duplicate_records == 0 else "Fail",
        "Pass" if negative_age == 0 else "Fail",
        "Pass" if negative_billing == 0 else "Fail",
        "Pass" if negative_stay == 0 else "Fail",
        "Pass" if invalid_dates == 0 else "Fail"
    ]
})

display(validation_report)
```

```python
# ==========================================
# INSPECT NEGATIVE BILLING AMOUNTS
# ==========================================

negative_billing_records = df[df["Billing Amount"] < 0]

display(negative_billing_records)
```

```python
# ==========================================
# FLAG NEGATIVE BILLING RECORDS
# ==========================================

df["Negative_Billing_Flag"] = df["Billing Amount"] < 0

display(
    df[[
        "Billing Amount",
        "Negative_Billing_Flag"
    ]].head(10)
)
```

```python
# ==========================================
# UPDATE VALIDATION REPORT
# ==========================================

missing_values = df.isnull().sum().sum()
duplicate_records = df.duplicated().sum()
negative_age = (df["Age"] < 0).sum()
negative_billing = (df["Billing Amount"] < 0).sum()
negative_stay = (df["Length_of_Stay"] < 0).sum()
invalid_dates = (
    df["Discharge Date"] <
    df["Date of Admission"]
).sum()

validation_report = pd.DataFrame({
    "Validation Check": [
        "Missing Values",
        "Duplicate Records",
        "Negative Age",
        "Negative Billing Amount",
        "Negative Length of Stay",
        "Invalid Dates"
    ],
    "Result": [
        missing_values,
        duplicate_records,
        negative_age,
        negative_billing,
        negative_stay,
        invalid_dates
    ],
    "Status": [
        "Pass" if missing_values == 0 else "Fail",
        "Pass" if duplicate_records == 0 else "Fail",
        "Pass" if negative_age == 0 else "Fail",
        "Review" if negative_billing > 0 else "Pass",
        "Pass" if negative_stay == 0 else "Fail",
        "Pass" if invalid_dates == 0 else "Fail"
    ]
})

display(validation_report)
```
